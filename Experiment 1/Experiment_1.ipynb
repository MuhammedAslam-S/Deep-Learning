```python
# 1)  Create a vector, scalar, matrix and tensor with values of your choice using tf.constant().


import tensorflow as tf

scalar = tf.constant(7)
print("Scalar:", scalar)

vector = tf.constant([1, 2, 3])
print("Vector:", vector)

matrix = tf.constant([[1, 2], [3, 4]])
print("Matrix:\n", matrix)

tensor = tf.constant([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print("Tensor:\n", tensor)
```

```python
#  2) Find the shape, rank and size of the tensors (which you created for Q1).


def describe_tensor(t):
    print("Tensor:", t.numpy())
    print("Shape:", t.shape)
    print("Rank:", tf.rank(t).numpy())
    print("Size:", tf.size(t).numpy())
    print()

describe_tensor(scalar)
describe_tensor(vector)
describe_tensor(matrix)
describe_tensor(tensor)
```

```python
# 3) Create two tensors containing random values between 0 and 1 with shape [5, 300].

tensor1 = tf.random.uniform(shape=[5, 300], minval=0.0, maxval=1.0)
tensor2 = tf.random.uniform(shape=[5, 300], minval=0.0, maxval=1.0)

print("Tensor 1 :\n",tensor1)
print("Tensor 2 :\n",tensor2)
```

```python
# 4)  Multiply the two tensors (which you created for Question 3) using matrix multiplication.

# Transpose tensor2 to make shapes compatible
tensor2_T = tf.transpose(tensor2)

result = tf.matmul(tensor1, tensor2_T)

print("Result :", result)
```

```python
# 5) Multiply the two tensors (which you created for Question 3) using dot product.

dot_product = tf.reduce_sum(tf.multiply(tensor1, tensor2), axis=1)

print("Dot Product :", dot_product.numpy())
```

```python
# 6) Create a tensor with random values between 0 and 1 with shape [224, 224, 3].

new_tensor = tf.random.uniform(shape=[224, 224, 3], minval=0.0, maxval=1.0)

print("New Tensor :", new_tensor)
```

```python
# 7) Find the min and max values of the tensor (which you created for Question 6)

new_tensor_min = tf.reduce_min(new_tensor)
new_tensor_max = tf.reduce_max(new_tensor)

print("Min Value : ",new_tensor_min)
print("Max Value : ",new_tensor_max)
```

```python
# 8) Created a tensor with random values of shape [1, 224, 224, 3] then squeeze it to change the shape to [224, 224, 3].

input_tensor = tf.random.uniform(shape=[1, 224, 224, 3], minval=0.0, maxval=1.0)

squeezed_tensor = tf.squeeze(input_tensor)

print("Original :", input_tensor)
print("Squeezed :", squeezed_tensor)
```

```python
# 9) Create a tensor with shape [10] using your own choice of values, then find the index which has the maximum value.

tensor9 = tf.constant([4, 8, 15, 16, 23, 42, 7, 3, 1, 9])

max_index = tf.argmax(tensor9)

print("Tensor:", tensor9.numpy())
print("Index of maximum value:", max_index.numpy())
```

```python
# 10) One-hot encode the tensor you created in Q.9

depth = tf.reduce_max(tensor9) + 1

one_hot_encoded = tf.one_hot(tensor9, depth=depth)

# Output
print("One-hot encoded tensor (first row):", one_hot_encoded[0].numpy())
```

```python
import tensorflow as tf

# Suppose image_tensor is your float tensor with shape [224, 224, 3]
image_tensor = tf.random.uniform(shape=[224, 224, 3], minval=0.0, maxval=1.0)

# Convert float [0,1] to uint8 [0,255]
image_uint8 = tf.image.convert_image_dtype(image_tensor, dtype=tf.uint8)

print("Original dtype:", image_tensor.dtype)
print("Converted dtype:", image_uint8.dtype)
print("Shape:", image_uint8.shape)
print(image_uint8)
```

```python
import matplotlib.pyplot as plt

plt.imshow(image_tensor.numpy())  # values between 0-1 work fine here
plt.axis('off')
plt.title("Random RGB Image")
plt.show()
```
